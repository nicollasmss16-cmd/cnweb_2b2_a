# formula da velocidade


```
def calculaVelMedia(deltaS, deltaT):
  velmedia = (deltaS/deltaT)
  return velmedia
```

## Teste Serviço Nuvem: velocidade media

```
print("Vm=", calculaVelMedia(30, 15))
```

# Fórmula da aceleração

```
def calcula_raceleracao(vi, vf, t):
  return (vf - vi) / t
```

```
resultado = calcula_raceleracao(0, 20, 5)
print(f"resultado: {resultado} m/s²")
```

#Função horária do deslocamento

```
def calculaDesloc(inicialS, deltaT, v):
  d = inicialS + (v*deltaT)
  return d
```

#Testes de Função : Horária de Deslocamento

```
print("Deslocamento= ",calculaDesloc(0, 15, 30))
```
